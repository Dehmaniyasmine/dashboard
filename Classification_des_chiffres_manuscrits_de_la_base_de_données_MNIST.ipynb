```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
from sklearn.model_selection import train_test_split
import numpy as np
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

# 1. Load and Preprocess Data
(x_train, y_train), (x_test, y_test) = keras.datasets.mnist.load_data()

# Concatenate and split
X = np.concatenate([x_train, x_test])
y = np.concatenate([y_train, y_test])
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Flatten and normalize
X_train = X_train.reshape(X_train.shape[0], -1).astype('float32') / 255
X_test = X_test.reshape(X_test.shape[0], -1).astype('float32') / 255

# One-hot encode labels
y_train = keras.utils.to_categorical(y_train, num_classes=10)
y_test = keras.utils.to_categorical(y_test, num_classes=10)

# 2. Define the MLP Model
model = keras.Sequential(
    [
        keras.Input(shape=(784,)),  # Input layer (784 features for flattened images)
        layers.Dense(512, activation="relu"),  # Hidden layer with 512 neurons
        layers.Dropout(0.2),  # Dropout for regularization
        layers.Dense(256, activation="relu"),  # Hidden layer with 256 neurons
        layers.Dropout(0.2),  # Dropout for regularization
        layers.Dense(10, activation="softmax"),  # Output layer with 10 neurons (for 10 digits)
    ]
)

# 3. Compile the Model
model.compile(loss="categorical_crossentropy", optimizer="adam", metrics=["accuracy"])

# 4. Train the Model
model.fit(X_train, y_train, batch_size=128, epochs=10, validation_split=0.1)

# 5. Predict Labels
y_pred_probs = model.predict(X_test)
y_pred_labels = np.argmax(y_pred_probs, axis=1)

# 6. Evaluate the Model
# Convert y_test to multiclass format for accuracy calculation
y_test_multiclass = np.argmax(y_test, axis=1)

accuracy = accuracy_score(y_test_multiclass, y_pred_labels)
print(f"Accuracy: {accuracy}")

print(classification_report(y_test_multiclass, y_pred_labels))
print(confusion_matrix(y_test_multiclass, y_pred_labels))
```
